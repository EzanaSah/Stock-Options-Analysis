```python
# Import Required Packages 
import requests, json, os
import numpy as np 
import pandas as pd
import opstrat as op
import yfinance as yf
import matplotlib.pyplot as plt
```

### Extract Options Pricing Data Using the Yahoo Finance Package 

The type of data that can be extracted can be found in the documentation here: https://pypi.org/project/yfinance/

```python
# Extract data for a specific Exchange Traded Fund (ETF)
etf = yf.Ticker("AAPL")

# Look at ETF info
etf_hist = etf.history(period='max')

# Plot the ETF Pricing History
etf_hist['Close'].plot()
plt.xlabel("Year")
plt.ylabel("Price (USD)")
plt.title(etf)
```

```python
print("S&P 500 Descriptive Statistics")
print(etf_hist['Close'].describe())
```

### Options Definitions

There are two types of options contracts: 

Call = the right to buy/sell an underlying asset at a certain price (strike price) at a certain date (expiration date).

Put = the obligation to buy/sell an underlying asset at a certain (strike price) at a certain date (expiration date).

There are many strategies that combine several options and create various risk/reward profiles.

```python
# View the Option Expiration Dates for the Underlying Asset 
expiration_dates = etf.options
```

```python
# View the Option Chain for the second Expiration Date available
opt_chain = etf.option_chain(expiration_dates[1])
opt_chain.calls
```

### Types of Single Options Contracts

There are four different types of options you can buy/sell. 
These contracts include 

Buying Call = Long Call = Buying right to Buy at strike  
Selling Call = Short Call = Selling right to Buy at strike  
Buying Put = Long Put = Buying obligation to Buy at strike  
Selling Put = Short Put = Selling obligation to Buy at strike

```python
# Option Seller, receives premium of $12.50 for selling a put option 
# Strike price of $460 & stock is trading at $460 (spot price).
op.single_plotter(spot=460, strike=460, 
                  op_type='c', tr_type='b',
                  op_pr=12.5)
```

```python
# Option Seller, receives premium of $12.50 for buying a put option 
# Strike price of $460 & stock is trading at $460 (spot price).
op.single_plotter(spot=460, strike=460, 
                  op_type='c', tr_type='s',
                  op_pr=12.5)
```
